```python
from PIL import Image
import numpy as np
import math
import scipy.misc

def prewit(img,kernel_prewitx,kernel_prewity):
    rows,cols=img.shape
    
    prewit_image=np.zeros([rows,cols])
    for i in np.arange(1,rows-1):
         for j in np.arange(1,cols-1):
                Gx=np.sum(kernel_prewitx*img[i-1:i+2,j-1:j+2])
                Gy=np.sum(kernel_prewity*img[i-1:i+2,j-1:j+2])
                prewit_image[i,j]=((Gx*Gx)+(Gy*Gy))**0.5
                print(np.arctan2(Gx,Gy))
    return prewit_image

img = Image.open('apple.jpg')
img2array = np.array(img)   
kernel_prewitx= np.array(([-1, 0, 1],[-1, 0, 1],[-1, 0, 1]), dtype="int")
kernel_prewity= np.array(([1, 1, 1],[0, 0, 0],[-1, -1, -1]), dtype="int")
prewit_image=prewit(img2array,kernel_prewitx,kernel_prewity)
scipy.misc.imsave('prewit_image.jpg', prewit_image)
final_output=Image.fromarray(prewit_image)
final_output.show()
```
